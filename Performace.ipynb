```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv('/content/compare_csv.csv')  # Replace with your file path

# 1. Bar Chart - Performance Comparison across Methods
def plot_bar_chart(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    data.plot(kind='bar', x='Method', ax=ax)
    plt.title('Performance Metrics Comparison', size=15)
    plt.ylabel('Performance')
    plt.xticks(rotation=45)
    plt.show()

# 2. Heatmap - Correlation between Metrics
def plot_heatmap(data):
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.heatmap(data.iloc[:, 1:].corr(), annot=True, cmap="coolwarm", ax=ax)
    plt.title('Correlation Heatmap of Performance Metrics', size=15)
    plt.show()

# 3. Box Plot - Distribution of Metrics across Methods
def plot_boxplot(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.boxplot(data=data.melt(id_vars='Method', var_name='Metric', value_name='Score'), x='Metric', y='Score')
    plt.title('Distribution of Metrics across Methods', size=15)
    plt.show()

# 4. Line Chart - Trend of Metrics across Methods
def plot_line_chart(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    for col in data.columns[1:]:
        ax.plot(data['Method'], data[col], marker='o', label=col)
    plt.title('Performance Metrics Trends by Method', size=15)
    plt.ylabel('Performance')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

# 5. Pair Plot - Pairwise Relationships of Metrics
def plot_pairplot(data):
    numeric_data = data.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
    sns.pairplot(numeric_data, kind="scatter", diag_kind="kde")
    plt.suptitle('Pairwise Relationships of Performance Metrics', size=15, y=1.02)
    plt.show()

# 6. Radar Chart - Multi-Metric Visualization for Each Method
def plot_radar_chart(data):
    categories = list(data.columns[1:])
    methods = data['Method']
    values = data.iloc[:, 1:].values
    N = len(categories)

    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]

    for i, method in enumerate(methods):
        values_for_method = values[i].tolist()
        values_for_method += values_for_method[:1]
        ax.plot(angles, values_for_method, linewidth=2, linestyle='solid', label=method)
        ax.fill(angles, values_for_method, alpha=0.25)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    plt.title('Radar Chart of Methods', size=15, color='blue', y=1.1)
    ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.show()

# Example usage:
plot_bar_chart(data)
plot_heatmap(data)
plot_boxplot(data)
plot_line_chart(data)
plot_pairplot(data)
plot_radar_chart(data)
```

```python
def plot_violin(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.violinplot(data=data.melt(id_vars='Method', var_name='Metric', value_name='Score'), x='Metric', y='Score')
    plt.title('Violin Plot of Metrics Distribution by Method', size=15)
    plt.show()

# 2. Swarm Plot - Individual data points distribution
def plot_swarm(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.swarmplot(data=data.melt(id_vars='Method', var_name='Metric', value_name='Score'), x='Metric', y='Score')
    plt.title('Swarm Plot of Metrics by Method', size=15)
    plt.show()

# 3. Strip Plot - Shows all data points with jitter
def plot_strip(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.stripplot(data=data.melt(id_vars='Method', var_name='Metric', value_name='Score'), x='Metric', y='Score', jitter=True)
    plt.title('Strip Plot with Jitter for Each Metric', size=15)
    plt.show()

# 4. Histogram - Frequency distribution of each metric
def plot_histogram(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    data.iloc[:, 1:].plot(kind='hist', bins=10, alpha=0.6, ax=ax)
    plt.title('Histogram of Performance Metrics', size=15)
    plt.ylabel('Frequency')
    plt.show()

# 5. Density Plot - Smooth kernel density estimate for each metric
def plot_density(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    data.iloc[:, 1:].plot(kind='kde', ax=ax)
    plt.title('Density Plot of Performance Metrics', size=15)
    plt.ylabel('Density')
    plt.show()

# 6. Facet Grid Plot - Multi-plot grid to visualize metrics
def plot_facet_grid(data):
    melted_data = data.melt(id_vars='Method', var_name='Metric', value_name='Score')
    g = sns.FacetGrid(melted_data, col="Metric", col_wrap=3, height=4)
    g.map(sns.barplot, "Method", "Score", palette="coolwarm")
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle('Facet Grid of Metrics across Methods', size=16)
    plt.show()

# Example usage:
plot_violin(data)
plot_swarm(data)
plot_strip(data)
plot_histogram(data)
plot_density(data)
plot_facet_grid(data)
```

```python
def plot_violin_chart(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.violinplot(data=data.melt(id_vars='Method', var_name='Metric', value_name='Score'), x='Metric', y='Score', hue='Method', split=True)
    plt.title('Distribution of Metrics across Methods (Violin Plot)', size=15)
    plt.show()

# 2. Swarm Plot - Show each point in the dataset for Metrics across Methods
def plot_swarm_chart(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.swarmplot(data=data.melt(id_vars='Method', var_name='Metric', value_name='Score'), x='Metric', y='Score', hue='Method', dodge=True)
    plt.title('Distribution of Metrics across Methods (Swarm Plot)', size=15)
    plt.show()

# 3. Facet Grid - Individual plots of each metric across Methods
def plot_facet_grid(data):
    melted_data = data.melt(id_vars='Method', var_name='Metric', value_name='Score')
    g = sns.FacetGrid(melted_data, col="Metric", col_wrap=3, height=4)
    g.map(sns.barplot, "Method", "Score")
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle('Facet Grid: Performance Metrics by Method', size=15)
    plt.show()

# 4. Stacked Area Plot - Cumulative Metric Comparison across Methods
def plot_stacked_area_chart(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    data.set_index('Method').T.plot(kind='area', stacked=True, ax=ax, cmap='tab20')
    plt.title('Cumulative Performance Metrics Comparison (Stacked Area Plot)', size=15)
    plt.ylabel('Metric Score')
    plt.show()

# 5. 3D Surface Plot - Relationship between three metrics in 3D
def plot_3d_surface_chart(data):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    X = np.arange(len(data['Method']))
    Y = np.arange(len(data.columns[1:]))
    X, Y = np.meshgrid(X, Y)
    Z = data.iloc[:, 1:].values.T

    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_xticks(np.arange(len(data['Method'])))
    ax.set_xticklabels(data['Method'], rotation=45, ha='right')
    ax.set_yticks(np.arange(len(data.columns[1:])))
    ax.set_yticklabels(data.columns[1:])

    ax.set_title('3D Surface Plot of Metrics', size=15)
    ax.set_zlabel('Performance Score')
    plt.show()

# 6. Density Plot (KDE Plot) - Distribution of each Metric
def plot_density_chart(data):
    fig, ax = plt.subplots(figsize=(10, 6))
    for col in data.columns[1:]:
        sns.kdeplot(data[col], label=col, ax=ax, shade=True)
    plt.title('Density Plot of Metrics (KDE Plot)', size=15)
    plt.xlabel('Performance')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

# Example usage:
plot_violin_chart(data)
plot_swarm_chart(data)
plot_facet_grid(data)
plot_stacked_area_chart(data)
plot_3d_surface_chart(data)
plot_density_chart(data)
```
